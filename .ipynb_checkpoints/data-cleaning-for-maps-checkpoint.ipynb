```python
# import packages
import pandas as pd
import re
import datetime
import missingno as msno
import numpy as np
import math 
import plotly.express as px
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')
```

```python
plc = pd.read_csv("cleaned_fc_data_2015_2019.csv")
indvsl = pd.read_csv("cleaned_fc_name_data_2015_2019.csv")
```

```python
# plc_noOb = plc[plc.circumstance != 'observed']
plc_stopped = plc[plc.circumstance == 'stopped']
indvsl_stopped = indvsl[indvsl.circumstance == 'stopped']
frisked_involved = plc_stopped[plc_stopped.person_involved_in_frisked_or_searched == "y"].reset_index(drop=True)
```

```python
# demographic data from https://statisticalatlas.com/zip/02124/Race-and-Ethnicity (考慮別的：https://www.unitedstateszipcodes.org/02108/#stats)
demo = pd.read_csv("data/zip_pop.csv")
remove = ["0ma", "0nan", "00000"]
demo["zip"] = [e for e in plc_stopped.zip.unique().tolist() if e not in remove]
```

```python
raceStopZipWithoutNan = indvsl_stopped[(indvsl_stopped.zip != "0nan") & (indvsl_stopped.zip != "00") & (indvsl_stopped.zip != "00000")]
raceGrouped_zip_fc = raceStopZipWithoutNan.groupby(["zip", "race"]).count()["contact_date"]
raceGrouped_zip_fc = raceGrouped_zip_fc.reset_index()
raceGrouped_zip_fc["num_of_stops"] = raceGrouped_zip_fc["contact_date"] 
raceGrouped_zip_fc.drop(["contact_date"], axis=1, inplace=True)
raceGrouped_zip_fc = raceGrouped_zip_fc[(raceGrouped_zip_fc.race=="black") | (raceGrouped_zip_fc.race=="white")]
raceGrouped_zip_fc.reset_index(inplace=True, drop=True)
```

```python
raceGrouped_zip_fc_black = raceGrouped_zip_fc[raceGrouped_zip_fc.race=="black"].rename(columns={"num_of_stops": "black_num_of_stops"}).reset_index(drop=True)
raceGrouped_zip_fc_white = raceGrouped_zip_fc[raceGrouped_zip_fc.race=="white"].rename(columns={"num_of_stops": "white_num_of_stops"}).reset_index(drop=True)
raceGrouped_zip_fc_black.drop(["race"], axis=1, inplace=True)
raceGrouped_zip_fc_white.drop(["race"], axis=1, inplace=True)
```

```python
stopZipWithoutNan = indvsl_stopped[(indvsl_stopped.zip != "0nan") & (indvsl_stopped.zip != "00") & (indvsl_stopped.zip != "00000")]
grouped_zip_fc = stopZipWithoutNan.groupby(["zip"]).count()["contact_date"]
grouped_zip_fc = grouped_zip_fc.to_frame()
grouped_zip_fc.reset_index(level=0, inplace=True)
grouped_zip_fc["num_of_stops"] = grouped_zip_fc["contact_date"] 
grouped_zip_fc.drop(["contact_date"], axis=1, inplace=True)
merged = grouped_zip_fc.merge(demo, left_on="zip", right_on="zip", how="outer")
merged = merged.merge(raceGrouped_zip_fc_black, left_on="zip", right_on="zip", how="outer")
merged = merged.merge(raceGrouped_zip_fc_white, left_on="zip", right_on="zip", how="outer")
merged["stops_per"] = merged["num_of_stops"]/merged["population"]
merged.black_num_of_stops = merged.black_num_of_stops.apply(lambda x: 0 if math.isnan(x) else x)
merged.white_num_of_stops = merged.white_num_of_stops.apply(lambda x: 0 if math.isnan(x) else x)
merged["per_black_stopped_within_blacks"] = merged["black_num_of_stops"]/merged["black_pop"]
merged["per_white_stopped_within_whites"] = merged["white_num_of_stops"]/merged["white_pop"]
merged["per_stopped_black_to_white"] = merged["per_black_stopped_within_blacks"]-merged["per_white_stopped_within_whites"]
```

```python
# shape file: https://bostonopendata-boston.opendata.arcgis.com/datasets/53ea466a189b4f43b3dfb7b38fa7f3b6_1
shape = pd.read_csv("data/ZIP_Codes_shapefile.csv")
shape["zip"] = shape.ZIP5.apply(lambda x: "0"+str(x))
shape_with_stops = shape.merge(merged, left_on="zip", right_on="zip", how="outer")
```

```python
shape_with_stops.sort_values(by=["stops_per"], ascending=False)
```
